# Exploratory Data Analysis

```python
df.columns
```

```python
ax = sns.countplot(x='Size', data=df)
```

```python
ax = sns.countplot(x='Size', data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

# From the above graph you can see the most of the people buys M-size

```python
#groupby 
df.groupby(['Size'], as_index =False)['Quantity'].sum().sort_values(by='Quantity' ,ascending= False)
```

```python
S_qty = df.groupby(['Size'], as_index =False)['Quantity'].sum().sort_values(by='Quantity' ,ascending= False)

sns.barplot(x='Size' ,y='Quantity' ,data=S_qty)
```

# From the above graph the most of the quantity buys M-size in the sale

```python
#histogram
df['Size'].hist()
```

```python
df['Category']=df['Category'].astype(str)
column_data = df['Category']
plt.figure(figsize=(10,5))
plt.hist(column_data ,bins=20,edgecolor ='Black')
plt.xticks(rotation=90)
plt.show()
```

```python
#checking B2B data by using pie chart 
```

```python
ax = sns.countplot(x='B2B', data=df)
B2B_Check =df['B2B'].value_counts()

#plot the pie chart
plt.plot(B2B_check ,labels=B2B_check.Index ,autopct='%1.1f%')

#plt.axis('equal')
plt.show()
```

```python
#Prepare data for scatter plot
x_data=df['Category']
y_data=df['Size']

#plot the scatter plot
plt.scatter(x_data , y_data)
plt.xlabel('Category')
plt.ylabel('size')
plt.title('Scatter Plot')
plt.show()
```

```python
#plot count of cities by state
plt.figure(figsize=(12,6))
df['ship-state']=df['ship-state'].astype(str)
sns.countplot(data=df,x='ship-state')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution of State')
plt.xticks(rotation=90)
plt.show()
```

```python
#top_10_states
top_10_state = df['ship-state'].value_counts().head(10)
#plot count of cities by state
plt.figure(figsize=(12,6))
sns.countplot(data=df[df['ship-state'].isin(top_10_state.index)],x='ship-state')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution of State')
plt.xticks(rotation=90)
plt.show()
```

```python
#Most of the buyers are from the Maharastra state
```
